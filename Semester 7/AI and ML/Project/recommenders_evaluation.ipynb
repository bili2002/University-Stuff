```python
def test_recommender(recommender):
    user_games = [400, 49520]
    top_recs = recommender().recommend(user_games, k=5)
    top_recs = find_games_names(top_recs)

    print("User games:", find_games_names(user_games))
    print("Top 5 recommendations:")
    for game_id in top_recs:
        print(f" - {game_id}")
```
